## 3 Methods for plotting the graph

```python
# 1st method
fig = plt.figure()  #creates a figure
ax = fig.add_subplot()  #creates some axes
plt.show()
```

```python
#2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)  #add some data
plt.show()
```

## Two options for subplots

```python
# subplot option 1

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
```

```python
# subplot option 1

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one to use (pyplot vs matplotlib oo method)?

- When plotting something quikly, okay to use the pyplot method
- When plotting something advanced, use the OO method

```python
# creating a df for viewing patients having age > 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# pyplot method
over_50.plot(kind = 'scatter',
            x = 'age',
            y='chol',
            c='target')                # 'c' stands for color
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10,6))         ## figsize shows the width and height of the figure in inches
over_50.plot(kind = 'scatter',
            x = 'age',
            y='chol',
            c='target',
            ax= ax)

ax.set_xlim(45,100)             # set the limits for x axis
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))
# plot the data 
scatter = ax.scatter(x = over_50["age"],
                    y=over_50["chol"],
                     c=over_50["target"])

# customize the plot
ax.set(title="Heart disease and cholestrol levels",
      xlabel = "age",
      ylabel = "cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title = "target")

# add a horizontal line
ax.axhline(over_50["chol"].mean(),     # horizontal line at the mean of y axis
          linestyle = '--')
```

```python
over_50["target"].values
```

```python
over_50.head()
```

```python
# with 2 subplots
fig, ax = plt.subplots(nrows=2,
                      ncols=1,
                      figsize=(10,10))
#plot the 1st subplot
scatter = ax[0].scatter(x = over_50["age"],
                       y = over_50["chol"],
                       c = over_50["target"])
#customize ax[0]
ax[0].set(title = 'Heart disease and cholestrol levels',
         xlabel = 'age',
         ylabel = 'cholestrol')

#add a horizontal line over the mean 
ax[0].axhline(y = over_50["chol"].mean(),
          linestyle = '--')

#add a legend 
ax[0].legend(*scatter.legend_elements(), title = 'target')



#plot the 2nd subplot
scatter = ax[1].scatter(x = over_50["age"],
                       y = over_50["thalach"],
                       c = over_50["target"])
#customize ax[1]
ax[1].set(title = 'Heart disease and max heart rate',
         xlabel = 'age',
         ylabel = 'max heart rate')

#add a horizontal line over the mean 
ax[1].axhline(y = over_50["thalach"].mean(),
          linestyle = '--')

#add a legend 
ax[1].legend(*scatter.legend_elements(), title = 'target')


# add a title to the figure
fig.suptitle("Heart disease analyses", fontsize=16, fontweight='bold')
```

```python
# set the style
plt.style.use('ggplot')

# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))
# plot the data 
scatter = ax.scatter(x = over_50["age"],
                    y=over_50["chol"],
                     c=over_50["target"],
                    cmap = 'winter')       # this changes the color scheme

# customize the plot
ax.set(title="Heart disease and cholestrol levels",
      xlabel = "age",
      ylabel = "cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title = "target")

# add a horizontal line
ax.axhline(over_50["chol"].mean(),     # horizontal line at the mean of y axis
          linestyle = '--')
```

```python
# with 2 subplots
fig, ax = plt.subplots(nrows=2,
                      ncols=1,
                      figsize=(10,10))
#plot the 1st subplot
scatter = ax[0].scatter(x = over_50["age"],
                       y = over_50["chol"],
                       c = over_50["target"],
                       cmap = 'winter')
#customize ax[0]
ax[0].set(title = 'Heart disease and cholestrol levels',
         xlabel = 'age',
         ylabel = 'cholestrol')

#change ax[0] x axis limits
ax[0].set_xlim([50,80])

#add a horizontal line over the mean 
ax[0].axhline(y = over_50["chol"].mean(),
          linestyle = '--')

#add a legend 
ax[0].legend(*scatter.legend_elements(), title = 'target')



#plot the 2nd subplot
scatter = ax[1].scatter(x = over_50["age"],
                       y = over_50["thalach"],
                       c = over_50["target"],
                       cmap = 'winter')
#customize ax[1]
ax[1].set(title = 'Heart disease and max heart rate',
         xlabel = 'age',
         ylabel = 'max heart rate')

#change ax[1] x axis limits
ax[1].set_xlim([50,80])
ax[1].set_ylim([60,200])
#add a horizontal line over the mean 
ax[1].axhline(y = over_50["thalach"].mean(),
          linestyle = '--')

#add a legend 
ax[1].legend(*scatter.legend_elements(), title = 'target')


# add a title to the figure
fig.suptitle("Heart disease analyses", fontsize=16, fontweight='bold')
```
